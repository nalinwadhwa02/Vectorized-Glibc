```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(filenames):
    for files in filenames:
        filedata = []
        with open(files, "r") as f:
            filedata = f.readlines()
        df = []
        for l in range(1,len(filedata)):
            line = filedata[l]
            line = [float(l.replace(" ","")) for l in line.strip().split(",")]
            df.append(line)
        df = np.array(df)
        df = df[df[:,0].argsort()]
        print(df)
        labels = filedata[0].strip().split(",")[1:]
        # plt.plot(df[:,0], df[:,1], "-o")
        # plt.plot(df[:,0], df[:,2], "-o")
        # plt.plot(df[:,0], df[:,3], "-o")
        # plt.xlabel("length of input (n) log scale")
        # plt.ylabel("time taken in ms log scale")
        # plt.legend(labels)
        # plt.xscale('log')
        # plt.yscale('log')
        # print(files)
        # plt.title(files.strip().split("/")[-1].split("_")[0]+ " benchmark")
        # plt.show()
        plt.plot(df[:,0], df[:,1], "-o")
        plt.plot(df[:,0], df[:,2], "-o")
        plt.legend(labels[:-1])
        plt.xlabel("Position of character in string (n) log scale")
        plt.ylabel("time taken in ms log scale")
        plt.xscale('log')
        plt.yscale('log')
        plt.title(files.strip().split("/")[-1].split("_")[0]+ " benchmark")
        plt.show()
```

```python
filenames = ["benchmarks/strchr_resultfile"];
plot(filenames)
```
